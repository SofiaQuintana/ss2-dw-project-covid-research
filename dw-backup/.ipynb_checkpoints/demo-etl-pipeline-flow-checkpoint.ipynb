# Demo — Trazabilidad ETL: Bronze → Silver → Gold

Tomamos **registros concretos** de tres fuentes distintas y rastreamos campo a campo qué le pasa a cada valor
en cada capa de transformación.

| Capa | Propósito | Tecnología |
|------|-----------|------------|
| **Bronze** | Ingestión cruda — datos tal como llegan de la fuente | Databricks Delta Lake · `sandbox_bronze_ss2` |
| **Silver** | Limpieza, tipado y normalización | Databricks Delta Lake · `stage_silver_ss2` |
| **Gold** | Modelo dimensional Galaxy Schema | Databricks + PostgreSQL local · `gold_ss2` |

> **Nota:** Bronze y Silver viven en Databricks (no en el PostgreSQL local). Los registros de esas capas
> se muestran tal como quedaron guardados en Delta Lake; Gold se consulta en vivo desde PostgreSQL.

````python
import os
from pathlib import Path
import psycopg
import pandas as pd
from IPython.display import display, HTML
from dotenv import load_dotenv

load_dotenv(Path('.env'))

CONN = dict(
    host='localhost',
    port=int(os.environ['PG_PORT']),
    dbname=os.environ['PG_DB'],
    user=os.environ['PG_USER'],
    password=os.environ['PG_PASSWORD'],
)

conn = psycopg.connect(**CONN)
print('Conexión al Gold (PostgreSQL) OK —', conn.info.server_version)
````

````python
def mostrar_evolucion(registros: dict, titulo: str = ""):
    """Tabla comparativa Bronze → Silver → Gold con colores por tipo de cambio."""
    bronze = registros.get('Bronze', {})
    silver = registros.get('Silver', {})
    gold   = registros.get('Gold',   {})

    all_fields = list(bronze.keys())
    for k in silver:
        if k not in all_fields:
            all_fields.append(k)
    for k in gold:
        if k not in all_fields:
            all_fields.append(k)

    def estilo(layer_dict, field, prev_dict):
        if field not in layer_dict:
            return 'background:#fafafa;color:#ccc'
        if field not in prev_dict:
            return 'background:#d4edda'          # verde: campo nuevo
        if str(layer_dict[field]) != str(prev_dict.get(field, '')):
            return 'background:#fff3cd'          # amarillo: valor cambió
        return 'background:#ffffff'              # blanco: sin cambios

    def fmt(v):
        if v == '—':
            return '<span style="color:#ccc">—</span>'
        return f'<span>{v}</span>'

    rows = []
    for field in all_fields:
        b_val = bronze.get(field, '—')
        s_val = silver.get(field, '—')
        g_val = gold.get(field,   '—')
        b_bg  = 'background:#e3f2fd' if field in bronze else 'background:#fafafa;color:#ccc'
        s_bg  = estilo(silver, field, bronze)
        g_bg  = estilo(gold,   field, silver)
        rows.append(
            f'<tr>'
            f'<td style="font-weight:600;padding:6px 14px;border:1px solid #e0e0e0;font-size:13px;white-space:nowrap">{field}</td>'
            f'<td style="{b_bg};padding:6px 14px;border:1px solid #e0e0e0;font-family:monospace;font-size:12px">{fmt(b_val)}</td>'
            f'<td style="{s_bg};padding:6px 14px;border:1px solid #e0e0e0;font-family:monospace;font-size:12px">{fmt(s_val)}</td>'
            f'<td style="{g_bg};padding:6px 14px;border:1px solid #e0e0e0;font-family:monospace;font-size:12px">{fmt(g_val)}</td>'
            f'</tr>'
        )

    legend = (
        '<div style="margin:10px 0 16px;font-size:12px;display:flex;gap:12px;flex-wrap:wrap">'
        '<span style="background:#d4edda;padding:3px 10px;border-radius:4px;border:1px solid #c3e6cb">🟢 Campo nuevo en esta capa</span>'
        '<span style="background:#fff3cd;padding:3px 10px;border-radius:4px;border:1px solid #ffeeba">🟡 Valor transformado</span>'
        '<span style="background:#e3f2fd;padding:3px 10px;border-radius:4px;border:1px solid #bbdefb">🔵 Dato crudo sin cambio</span>'
        '</div>'
    )
    header = (
        '<tr>'
        '<th style="background:#37474f;color:white;padding:8px 14px;border:1px solid #ccc;text-align:left">Campo</th>'
        '<th style="background:#6d4c41;color:white;padding:8px 14px;border:1px solid #ccc">🥉 BRONZE<br><small style="font-weight:normal">datos crudos</small></th>'
        '<th style="background:#546e7a;color:white;padding:8px 14px;border:1px solid #ccc">🥈 SILVER<br><small style="font-weight:normal">limpieza y normalización</small></th>'
        '<th style="background:#e65100;color:white;padding:8px 14px;border:1px solid #ccc">🥇 GOLD<br><small style="font-weight:normal">modelo dimensional</small></th>'
        '</tr>'
    )
    html = (
        '<div style="font-family:sans-serif;margin:24px 0">'
        f'<h3 style="color:#37474f;border-left:4px solid #1565c0;padding-left:12px;margin-bottom:6px">{titulo}</h3>'
        f'{legend}'
        '<div style="overflow-x:auto">'
        '<table style="border-collapse:collapse;min-width:700px;width:100%">'
        f'<thead>{header}</thead>'
        f'<tbody>{chr(10).join(rows)}</tbody>'
        '</table></div></div>'
    )
    display(HTML(html))
````

## Arquitectura del pipeline

```
Fuentes externas
  INE (S3)  ──────────────────────────────────────────────────────────┐
  MSPAS (Dropbox) ─────────────────────────────────────────────────── │
  WHO (Google Drive / Volume) ──────────────────────────────────────── │
                                                                       ▼
                                               ┌─────────────────────────────┐
                                               │  BRONZE (sandbox_bronze_ss2) │
                                               │  9 tablas Delta Lake         │
                                               │  + columnas de linaje        │
                                               └────────────┬────────────────┘
                                                            │  stage_*.ipynb
                                                            ▼
                                               ┌─────────────────────────────┐
                                               │  SILVER (stage_silver_ss2)  │
                                               │  9 tablas Delta Lake         │
                                               │  tipos correctos, CIE-10     │
                                               │  normalizado, sin subtotales │
                                               └────────────┬────────────────┘
                                                            │  gold_*.ipynb
                                                            ▼
                                 ┌──────────────────────────────────────────────┐
                                 │  GOLD (gold_ss2)  —  Galaxy Schema            │
                                 │  dim_tiempo  dim_geografia  dim_causa          │
                                 │  dim_demografia  dim_fuente                    │
                                 │  fact_defunciones  fact_morbilidad             │
                                 └────────────────────┬─────────────────────────┘
                                                      │  load-postgresql.py
                                                      ▼
                                         PostgreSQL local (este notebook)
```

## Ejemplo 1 — WHO Mortalidad: normalización de idioma, tipos y grupos etarios

Seguimos un registro de **muertes por COVID-19 en Guatemala, 2020, mujeres de 25-29 años**,
tal como entra desde el portal WHO Mortality Database y cómo queda en Gold.

````python
who_bronze = {
    # --- Campos de la fuente (CSV descargado del portal WHO) ---
    'year':                   '2020',                              # string en el CSV original
    'country':                'guatemala',                         # sin capitalizar
    'Gender':                 'Female',                            # etiqueta en inglés
    'age_group_code':         'Age25_29',                         # formato compacto
    'indicator_code':         'SA.D.COVID19',
    'indicator_name':         'Coronavirus disease (COVID-19)',    # en inglés
    'number':                 '142',                               # string, no entero
    'crude_death_rate':       '0.82',
    'age_std_death_rate':     '0.79',
    'pct_cause_total_deaths': '1.45',
    # --- Columnas de linaje añadidas por Bronze ---
    'ingestion_timestamp':    '2024-11-20 08:30:00',
    'source_system':          'WHO Mortality Database Portal',
    'source_file':            'WHO_Guatemala_2020.csv',
}

who_silver = {
    # --- Campos renombrados y tipados ---
    'anio':                      2020,                             # INT (era string '2020')
    'pais':                      'Guatemala',                      # initcap + limpieza
    'sexo':                      'F',                              # 'Female' → 'F'
    'grupo_etario':              '25-29',                          # 'Age25_29' → '25-29'
    'codigo_causa':              'SA.D.COVID19',                   # = indicator_code
    'nombre_causa':              'Enfermedad por coronavirus (COVID-19)',  # traducido al español
    'defunciones':               142,                              # BIGINT (era string)
    'tasa_mortalidad_x100k':     0.82,
    'tasa_estandarizada_x100k':  0.79,
    'pct_causa_total_muertes':   1.45,
    # --- Columnas de auditoría Silver ---
    'silver_processed_timestamp': '2024-11-20 09:15:44',
    'silver_job_run_id':          'run-who-gt-001',
    # (ingestion_timestamp y source_* se propagan sin cambio)
    'ingestion_timestamp':    '2024-11-20 08:30:00',
    'source_system':          'WHO Mortality Database Portal',
    'source_file':            'WHO_Guatemala_2020.csv',
}

who_gold = {
    # --- Foreign keys (surrogate keys calculadas con row_number()) ---
    'tiempo_sk':      9,      # dim_tiempo: anio=2020
    'geografia_sk':   1,      # dim_geografia: pais='Guatemala', nivel_geo='pais'
    'demografia_sk':  15,     # dim_demografia: sexo='F', grupo_etario_label='25-29'
    'causa_sk':       87,     # dim_causa: codigo_origen='SA.D.COVID19'
    'fuente_sk':      3,      # dim_fuente: source_system='WHO Mortality Database Portal'
    # --- Medidas desnormalizadas en el fact ---
    'total_defunciones':        142,
    'defunciones_hombres':      None,  # WHO no desagrega por sexo en el fact de defunciones
    'defunciones_mujeres':      None,
    'tasa_mortalidad_x100k':    0.82,
    'tasa_estandarizada_x100k': 0.79,
    # (anio, pais, sexo, etc. ya no están aquí — viven en las dimensiones)
}

mostrar_evolucion(
    {'Bronze': who_bronze, 'Silver': who_silver, 'Gold': who_gold},
    titulo='Ejemplo 1 — WHO: defunción COVID-19, Guatemala 2020, Mujeres 25-29'
)
````

````python
# Verificamos en Gold (PostgreSQL) que ese registro existe
SQL_WHO_GOLD = """
SELECT
    t.anio,
    t.periodo_covid,
    g.pais,
    g.nivel_geo,
    d.sexo,
    d.grupo_etario_label,
    c.codigo_origen,
    c.descripcion,
    fu.source_system,
    f.total_defunciones,
    f.tasa_mortalidad_x100k,
    f.tasa_estandarizada_x100k
FROM gold_ss2.fact_defunciones f
JOIN gold_ss2.dim_tiempo     t  ON f.tiempo_sk    = t.tiempo_sk
JOIN gold_ss2.dim_geografia  g  ON f.geografia_sk = g.geografia_sk
JOIN gold_ss2.dim_demografia d  ON f.demografia_sk = d.demografia_sk
JOIN gold_ss2.dim_causa      c  ON f.causa_sk     = c.causa_sk
JOIN gold_ss2.dim_fuente     fu ON f.fuente_sk    = fu.fuente_sk
WHERE t.anio      = 2020
  AND g.pais      = 'Guatemala'
  AND g.nivel_geo = 'pais'
  AND d.sexo      = 'F'
  AND fu.source_system ILIKE '%WHO%'
ORDER BY f.total_defunciones DESC
LIMIT 8
"""

df_who_gold = pd.read_sql(SQL_WHO_GOLD, conn)
print(f'Registros en Gold (Guatemala 2020, mujeres, fuente WHO): {len(df_who_gold)}')
df_who_gold
````

## Ejemplo 2 — INE Defunciones: normalización CIE-10 y filtrado de subtotales

El INE entrega los códigos CIE-10 con separadores irregulares (`J:18:0`).
Silver los limpia a formato estándar (`J180`). Además los archivos INE incluyen
filas de subtotal (`'Todas las causas'`) que se **eliminan** en Silver.

````python
ine_bronze = {
    # --- Fila real tal como viene del XLSX del INE ---
    'anio':           '2020',               # string (XLSX lo exporta así a veces)
    'sexo':           'Hombres',
    'edad':           '25-29',
    'codigo_cie_10':  'J:18:0',             # ← separadores irregulares
    'causa_de_muerte':'Neumonía, no especificada',
    'total':          '89',                 # string
    'hombres':        '89',
    'mujeres':        '0',
    # --- Linaje Bronze ---
    'ingestion_timestamp': '2024-11-20 07:55:12',
    'source_system':       'INE_DEFUNCIONES',
    'source_file':         's3://bucket-ss2/ine/defunciones_sexo_edad_2020.xlsx',
}

# Fila de subtotal que Silver va a eliminar
ine_bronze_subtotal = {
    'anio':           '2020',
    'sexo':           'Total',
    'edad':           'Todas las edades',
    'codigo_cie_10':  'Todas las causas',   # ← subtotal: ELIMINADO en Silver
    'causa_de_muerte':'Todas las causas',
    'total':          '42115',
    'hombres':        '22890',
    'mujeres':        '19225',
    'ingestion_timestamp': '2024-11-20 07:55:12',
    'source_system':       'INE_DEFUNCIONES',
    'source_file':         's3://bucket-ss2/ine/defunciones_sexo_edad_2020.xlsx',
}

ine_silver = {
    # --- Campos transformados ---
    'anio':           2020,                 # INT (era string)
    'sexo':           'Hombres',
    'edad':           '25-29',
    'cie_10_norm':    'J180',               # ← J:18:0 → J180 (sin separadores)
    'codigo_cie_10':  'J:18:0',             # original preservado
    'causa_de_muerte':'Neumonía, no especificada',
    'total':          89,                   # INT
    'hombres':        89,
    'mujeres':        0,
    # --- Auditoría Silver ---
    'silver_processed_timestamp': '2024-11-20 09:20:05',
    'silver_job_run_id':          'run-ine-edad-001',
    'ingestion_timestamp': '2024-11-20 07:55:12',
    'source_system':       'INE_DEFUNCIONES',
    'source_file':         's3://bucket-ss2/ine/defunciones_sexo_edad_2020.xlsx',
}

ine_gold = {
    # --- Surrogates (dimensiones) ---
    'tiempo_sk':     9,    # anio=2020
    'geografia_sk':  2,    # placeholder ('Guatemala','Sin desagregar') — INE-edad no tiene dept.
    'demografia_sk': 22,   # sexo='Hombres', grupo_etario_label='25-29'
    'causa_sk':      134,  # codigo_origen='J180'
    'fuente_sk':     1,    # INE_DEFUNCIONES, pipeline_name='ine_defunciones_sexo_edad'
    # --- Medidas ---
    'total_defunciones':    89,
    'defunciones_hombres':  89,
    'defunciones_mujeres':  0,
    'tasa_mortalidad_x100k':    None,  # INE-edad no provee tasas
    'tasa_estandarizada_x100k': None,
}

mostrar_evolucion(
    {'Bronze': ine_bronze, 'Silver': ine_silver, 'Gold': ine_gold},
    titulo='Ejemplo 2 — INE: defunción por neumonía (J18.0), Hombres 25-29, Guatemala 2020'
)

display(HTML(
    '<div style="background:#fdecea;border-left:4px solid #e53935;padding:10px 16px;'
    'border-radius:4px;font-family:sans-serif;font-size:13px;margin:10px 0">'
    '<strong>🗑️ Fila de subtotal ELIMINADA en Silver:</strong><br>'
    '<code>anio=2020 | sexo=Total | edad=Todas las edades | codigo_cie_10=<b>\'Todas las causas\'</b> | total=42115</code><br>'
    'Silver filtra todas las filas donde <code>causa_de_muerte IN (\'Todas las causas\', \'Otras causas\')</code>'
    '</div>'
))
````

````python
# Verificamos en Gold: defunciones INE con código J18x, año 2020, hombres
SQL_INE_GOLD = """
SELECT
    t.anio,
    t.periodo_covid,
    g.pais,
    g.departamento,
    d.sexo,
    d.grupo_etario_label,
    c.codigo_origen,
    c.descripcion,
    fu.pipeline_name,
    f.total_defunciones,
    f.defunciones_hombres,
    f.defunciones_mujeres
FROM gold_ss2.fact_defunciones f
JOIN gold_ss2.dim_tiempo     t  ON f.tiempo_sk    = t.tiempo_sk
JOIN gold_ss2.dim_geografia  g  ON f.geografia_sk = g.geografia_sk
JOIN gold_ss2.dim_demografia d  ON f.demografia_sk = d.demografia_sk
JOIN gold_ss2.dim_causa      c  ON f.causa_sk     = c.causa_sk
JOIN gold_ss2.dim_fuente     fu ON f.fuente_sk    = fu.fuente_sk
WHERE t.anio = 2020
  AND c.codigo_origen LIKE 'J18%'
  AND fu.pipeline_name ILIKE '%ine%edad%'
ORDER BY d.sexo, d.grupo_etario_label
LIMIT 10
"""

df_ine_gold = pd.read_sql(SQL_INE_GOLD, conn)
print(f'Registros en Gold (código J18x, INE-edad, 2020): {len(df_ine_gold)}')
df_ine_gold
````

## Ejemplo 3 — MSPAS Morbilidad: COALESCE de columnas y resolución de departamento

Los archivos MSPAS tienen dos columnas para casos según el período (`casos` / `cantidad`).
Silver los unifica con `COALESCE`. Además el nombre del departamento llega en mayúsculas
y se resuelve contra el catálogo oficial con tildes y casing correcto.

````python
mspas_bronze = {
    # --- CSV de MSPAS Dropbox ---
    'anio':                  2020.0,               # DOUBLE (Spark lo lee así desde CSV)
    'departamento':          'CIUDAD DE GUATEMALA', # mayúsculas sin tilde
    'municipio':             'Zona 1',
    'codigo_cie_10':         'J:06',               # separador irregular
    'causa_de_morbilidad':   'Infecciones agudas de las vías respiratorias superiores',
    'casos':                 None,                  # columna vacía en este período
    'cantidad':              1840,                  # columna alternativa con el dato
    'sexo':                  'F',
    # --- Linaje Bronze ---
    'ingestion_timestamp':   '2024-11-20 08:10:33',
    'source_system':         'MSPAS',
    'source_file':           'primeras_causas_morbilidad_2020.csv',
}

mspas_silver = {
    # --- Campos transformados ---
    'anio':                2020,            # INT (era 2020.0 double)
    'departamento_oficial':'Guatemala',      # resuelto contra catálogo ref_departamentos_gt
    'departamento':        'CIUDAD DE GUATEMALA',  # original preservado
    'municipio':           'Zona 1',
    'cie_10_norm':         'J06',           # 'J:06' → 'J06'
    'codigo_cie_10':       'J:06',          # original preservado
    'causa_de_morbilidad': 'Infecciones agudas de las vías respiratorias superiores',
    'casos_total':         1840,            # COALESCE(casos, cantidad) = 1840
    'casos':               None,
    'cantidad':            1840,
    'sexo':                'F',
    # --- Auditoría Silver ---
    'silver_processed_timestamp': '2024-11-20 09:30:18',
    'silver_job_run_id':          'run-mspas-001',
    'ingestion_timestamp':   '2024-11-20 08:10:33',
    'source_system':         'MSPAS',
    'source_file':           'primeras_causas_morbilidad_2020.csv',
}

mspas_gold = {
    # --- Surrogates ---
    'tiempo_sk':    9,    # anio=2020
    'geografia_sk': 48,   # pais='Guatemala', departamento='Guatemala', municipio='Zona 1'
    'demografia_sk':10,   # sexo='F', grupo_etario_label='Sin desagregar'
    'causa_sk':     55,   # codigo_origen='J06'
    'fuente_sk':    6,    # MSPAS, pipeline_name='primeras_causas_de_morbilidad'
    # --- Medida única de morbilidad ---
    'casos_total':  1840,
    # (casos y cantidad no pasan al fact — ya se consolidaron en Silver)
}

mostrar_evolucion(
    {'Bronze': mspas_bronze, 'Silver': mspas_silver, 'Gold': mspas_gold},
    titulo='Ejemplo 3 — MSPAS: morbilidad respiratoria (J06), Guatemala Zona 1, 2020, Mujeres'
)
````

````python
# Verificamos en Gold: morbilidad MSPAS código J0x, departamento Guatemala, 2020
SQL_MSPAS_GOLD = """
SELECT
    t.anio,
    t.periodo_covid,
    g.departamento,
    g.municipio,
    g.nivel_geo,
    d.sexo,
    c.codigo_origen,
    c.descripcion,
    fu.pipeline_name,
    f.casos_total
FROM gold_ss2.fact_morbilidad f
JOIN gold_ss2.dim_tiempo     t  ON f.tiempo_sk    = t.tiempo_sk
JOIN gold_ss2.dim_geografia  g  ON f.geografia_sk = g.geografia_sk
JOIN gold_ss2.dim_demografia d  ON f.demografia_sk = d.demografia_sk
JOIN gold_ss2.dim_causa      c  ON f.causa_sk     = c.causa_sk
JOIN gold_ss2.dim_fuente     fu ON f.fuente_sk    = fu.fuente_sk
WHERE t.anio           = 2020
  AND g.departamento   = 'Guatemala'
  AND c.codigo_origen LIKE 'J0%'
ORDER BY f.casos_total DESC
LIMIT 10
"""

df_mspas_gold = pd.read_sql(SQL_MSPAS_GOLD, conn)
print(f'Registros en Gold (MSPAS, J0x, depto Guatemala, 2020): {len(df_mspas_gold)}')
df_mspas_gold
````

## Resumen: ¿qué transforma cada capa?

````python
resumen = [
    {
        'Capa':        '🥉 BRONZE',
        'Propósito':   'Ingestión — copia fiel de la fuente',
        'Qué hace':    'Carga los archivos sin modificar + añade columnas de linaje (ingestion_timestamp, source_system, source_file)',
        'Limitaciones':'Tipos incorrectos · Códigos con separadores · Textos en inglés · Subtotales mezclados · Columnas de casos duplicadas',
    },
    {
        'Capa':        '🥈 SILVER',
        'Propósito':   'Limpieza y normalización',
        'Qué hace':    'Castea tipos (anio→INT, total→INT) · Normaliza CIE-10 (J:18:0 → J180) · Traduce etiquetas al español · '
                       'Filtra subtotales · Resuelve departamentos contra catálogo · COALESCE de columnas de casos · '
                       'Parsea grupos etarios (Age25_29 → 25-29) · Elimina duplicados exactos',
        'Limitaciones':'Sin modelo dimensional — tablas planas, sin surrogate keys, sin métricas consolidadas',
    },
    {
        'Capa':        '🥇 GOLD',
        'Propósito':   'Modelo dimensional — listo para análisis',
        'Qué hace':    'Galaxy Schema: 5 dimensiones compartidas + 2 tablas de hechos · '
                       'Surrogate keys con row_number() · Periodo COVID calculado (anio ≥ 2020) · '
                       'Unifica todas las fuentes en el mismo modelo · Deduplica por grain',
        'Limitaciones':'Las fuentes INE-edad e INE-geografía pueden contar las mismas muertes (filtrar por dim_fuente)',
    },
]

df_resumen = pd.DataFrame(resumen).set_index('Capa')

def color_row(row):
    capa = row.name
    if 'BRONZE' in capa:
        return ['background:#fff8f4'] * len(row)
    if 'SILVER' in capa:
        return ['background:#f4f6f8'] * len(row)
    return ['background:#fff8e1'] * len(row)

df_resumen.style.apply(color_row, axis=1).set_properties(**{
    'font-size': '12px',
    'text-align': 'left',
    'vertical-align': 'top',
    'padding': '8px 12px',
    'white-space': 'pre-wrap',
})
````
